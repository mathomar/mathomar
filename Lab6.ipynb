# Working with Strings in Pandas

## Variant 3

```python
import pandas as pd
import re

#1
file = pd.read_csv("books.csv")
file
```

## 2 )Create the columns with number of vowels in author's first name

```python
#2
vwl = file['Author'].str.extract(', (?P<first_name>\w*)')
vwl['vowels_in_fn'] = vwl.first_name.str.lower().str.count(r'[aeiou]')
vwl
```

## 3) Find authors with 3 vowels in row in their last names¶

```python
#3
vwl1 = file['Author'].str.extract('(?P<last_name>\w*), ')
vwl1['Vowels'] = vwl1.last_name.str.lower().str.count(r'[aeiou]')
vwl1.loc[vwl1['Vowels'] == 3]
```

## 4) Find books where there are one letter word in the title

```python
#4
Titl =file[['Title','Author']].copy()   
Titl['Letter word Title']=Titl['Title'].str.count(r'\s\w$')
data = Titl[(Titl['Letter word Title'] != 0)]
data
```

## 5) Create a column with 1 if there is number in the title and o otherwise

```python
Titl['Number in the title']=Titl['Title'].str.count(r'(\d+)')
Titl.head(148)
```

## 6) Create a new column with last word of books' titles

```python
#6
fil2 = file[['Title','Author']].copy()
fil2['last word']=fil2['Title'].str.findall(r'\b(\w+)\W*$')
fil2.head(10)
```

## 7) Find the books that were published by publisher with 2 words in its name

```python
#7

publisher=file[['Title','Publisher']].copy()
publisher['publisher with 2 words']=publisher['Publisher'].str.count(r'\w\s\w')
new = publisher[(publisher['publisher with 2 words'] == 1)]
new
```

## 8) Find authors with the 3rd ‘y’ in the last name

```python
#8
vwl2 = file['Author'].str.extract('(?P<last_name>\w*), (?P<first_name>\w*)')
vwl2['3rd ‘y’'] = vwl2['last_name'].str.count(r'^\w\w[Y|y]\w')
authrs = vwl2[(vwl2['3rd ‘y’'] == 1)]
authrs
```
