## Task 3 And Task 4:

* 3- Rename the following list of countries:"Republic of Korea": "South Korea","United States of America": "United States", "United Kingdom of Great Britain and Northern Ireland": "UnitedKingdom","China, Hong Kong Special Administrative Region": "Hong Kong"
* 4-There are also several countries with numbers and/or parenthesis in their name. Be sure to remove these, e.g. 'Bolivia (Plurinational State of)' should be ‘Bolivia’, ‘Switzerland17’ should be Switzerland’.

```python
data1["Country"] = data1["Country"].replace(["Republic of Korea"], "South Korea")
data1["Country"] = data1["Country"].replace(["United States of America"], "United states")
data1["Country"] = data1["Country"].replace(["United Kingdom of Great Britain and Northern Ireland"], "United kingdom")
data1["Country"] = data1["Country"].replace(["China, Hong Kong Special Administrative Region"], "Hong Kong")

def rename(R):
    res = []
    for E in R:
        if E.isdigit():
            continue
        elif E == "(":
            break
        else:
            res.append(E)
    return "".join(res).rstrip()

data1["Country"] = data1["Country"].apply(lambda x: rename(x))
data1.loc[data1["Country"].isin(["American Samoa", "South Korea", "Bolivia"])]
```

## Task8:

* Create a function to define what are the top 15 countries for average GDP over the last 10 years

```python
top = (x for x in range(2006, 2016))
def part8():
    output["avgGDP"] = output[top].mean(axis=1, skipna=True)
    return output["avgGDP"].sort_values(ascending=False)
print(part8())
```

## Task9:

* Create a function to define by how much had the GDP changed over the past 10 year for the country with the 5th largest average GDP

```python
def part9():
    display = list(output.sort_values("avgGDP", ascending=False).index.values)[4]
    return (display, output.loc[display][2015] - output.loc[display][2006])
```

## Task 12:

* Create a column that estimates the number of citable documents per person. What is the correlation between the number of citable documents per capita and the energy supply per capita? Use the .corr() method, (Pearson's correlation).

```python
def part12():
    output["CitPerCap"] =  output["Citable documents"] / output["Population"]
    return output["Energy Supply per Capita"].astype('float64').corr(output['CitPerCap'].astype('float64'), method="pearson")
```

## Task 13:

* Create a new column with a 1 if the country's % Renewable value is at or above the median for all countries in the top 15, and a 0 if the country's % Renewable value is below the median.

```python
def part13():
    time = output["% Renewable"].median(axis=0)
    output["part13"] = output.apply(lambda x: 1 if x["% Renewable"] >= time else 0, axis=1)
    output.sort_values(by='Rank')
    return output["part13"]
```

## Task 14:

* Use the following dictionary to group the Countries by Continent, then create a DataFrame that displays the sample size (the number of countries in each continent bin), and the sum, mean, and std deviation for the estimated population of each country.

```python
def part14():
    Dict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
    time = pd.DataFrame(columns = ['size', 'sum', 'mean', 'std'])
    output['Population'] = output['Energy Supply'] / output['Energy Supply per Capita']
    for elem, val in output.groupby(Dict):
        time.loc[elem] = [len(val), val['Population'].sum(),val['Population'].mean(),val['Population'].std()]
    time = time.astype({"size":int})
    return time

```

```python
part14()
```

## Task 15:

* Create a bubble chart showing % Renewable vs. Rank. The size of the bubble corresponds to the countries' 2015 GDP, and the color corresponds to the continent

```python
def part15():
    colors = ['blue', 'green', 'red'] * 5
    x, y = output["Rank"], output["% Renewable"]
    for i, text in enumerate(list(output.index)):
        plt.annotate(text, (x[i], y[i]))
    plt.scatter(x=x, y=y, s=output[2015] / 10000000000*7, alpha=0.4, c=colors)
    plt.xticks([i for i in range(0, len(list(output.index)) + 1)])
    plt.gcf().set_size_inches((15, 5))
    plt.show()
```
